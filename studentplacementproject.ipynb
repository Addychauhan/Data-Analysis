# Student Placement Project

```python
#Import required Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
#Loading the data
data=pd.read_csv("/content/updated_placement.csv")
```

```python
#Basic operations
data.shape
```

```python
data.info()
```

```python
data.head(5)
```

```python
data
```

```python
plt.scatter(data["cgpa"],data["package"])
plt.xlabel("cgpa")
plt.ylabel("package")
plt.title("Student Placement Record")
```

```python
student_cgpa=data["cgpa"]
```

```python
student_cgpa=data["cgpa"].values.reshape(-1,1)
student_cgpa
```

```python
student_package=data["package"]
```

```python
student_package
```

splitting the data

```python
x_train, x_test, y_train, y_test = train_test_split(student_cgpa, student_package, test_size = 0.2)
```

```python
plt.scatter(x_train,y_train,label="Training Dataset")
plt.scatter(x_test,y_test,label="Testing Dataset")
plt.xlabel("CGPA")
plt.ylabel("Package")
plt.title("Student Placement Record")
plt.legend()
```

```python
obj=LinearRegression()
```

```python
obj
```

```python
obj.fit(x_train,y_train)
```

```python
x_test
```

```python
y_pred=obj.predict(x_test)
y_pred
```

```python
y_test
```

```python
r2_score(y_test,y_pred)
```

```python
plt.scatter(x_train,y_train,label="Training Dataset")
plt.scatter(x_test,y_test,label="Testing Dataset")
plt.plot(x_test,y_pred,color="red",label="Predticted Line")
plt.xlabel("CGPA")
plt.ylabel("Package")
plt.title("Student Placement Record")
plt.legend()
```
